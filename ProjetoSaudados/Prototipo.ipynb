```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
import unicodedata
import requests

def remove_special_characters(text):
    text = ''.join(c for c in unicodedata.normalize('NFKD', text) if not unicodedata.combining(c))
    return text.upper()

def on_button_click():
    selected_extension = extension_combobox.get()
    url = url_entry.get()
    mensagem_label.config(text=f"Lendo arquivo {url}...")

    try:
        if selected_extension == ".csv":
            df = pd.read_csv(url, delimiter=determinar_delimitador(url), low_memory=False)
        elif selected_extension == ".xlsx":
            df = pd.read_excel(url, delimiter=determinar_delimitador(url), low_memory=False)
        elif selected_extension == ".json":
            df = pd.read_json(url, delimiter=determinar_delimitador(url), low_memory=False)

        # Exibir as colunas temporariamente
        colunas = list(df.columns)
        cleaned_columns = [remove_special_characters(coluna) for coluna in colunas]

        # Limpa as opções atuais e define as colunas como opções no ComboBox
        columns_combobox['values'] = tuple(cleaned_columns)

        mensagem_label.config(text="Arquivo carregado com sucesso!")
            
    except Exception as e:
        mensagem_label.config(text=f"Erro ao carregar arquivo: {e}")


def determinar_delimitador(url):
    r = requests.get(url)
    line = r.content.decode().splitlines()[0]
    delimiters = [',', ';', '\t']
    for delimiter in delimiters:
        if delimiter in line:
            return delimiter
    return ','

# Cria a janela
root = tk.Tk()
root.geometry("1000x600") 
root.minsize(1000, 600) 
root.maxsize(1000, 600) 
root.title("SAUDADOS")

#espaço
space = tk.Label(root, text= "-----------------------------------------------------------------------")
space.pack()

# Name Site
Name = tk.Label(root, text="SAUDADOS – CIÊNCIA DE DADOS E SAÚDE")
Name.pack()

# Entrada
label = tk.Label(root, text= "Olá, Bem Vindo!")
label.pack()

#espaço
space = tk.Label(root, text= "-----------------------------------------------------------------------")
space.pack()

# Label para extensão
extension_label = tk.Label(root, text="Selecione a Extensão:")
extension_label.pack()

# ComboBox para extensão
extension_options = [".csv", ".xlsx", ".json"]  # Opções de extensão
extension_combobox = ttk.Combobox(root, values=extension_options)
extension_combobox.pack()

# Url Arquivo
input_url = ttk.Label(root, text=f"URL do arquivo {extension_options}:")
url_entry = ttk.Entry(root, width=40)
input_url.pack()
url_entry.pack()

# Cria um botão
button = tk.Button(root, text="Ler Arquivo", command=on_button_click)
button.pack()

# Rótulo para exibir colunas temporárias
columns_label = tk.Label(root, text="")
columns_label.pack()

# Rótulo para exibir mensagens de status
mensagem_label = tk.Label(root, text="")
mensagem_label.pack()

#Label colunas
label = tk.Label(root, text="Selecione a Coluna:")
label.pack()

# ComboBox para Colunas do arquivo (inicialmente vazio)
columns_combobox = ttk.Combobox(root)
columns_combobox.pack()

# Roda o loop principal
root.mainloop()
```
